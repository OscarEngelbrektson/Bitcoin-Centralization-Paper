# Concentration of mining power - ABM

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
miner_count = 1000
miners = np.full(miner_count,150)
winners = []
inices = np.arange(miner_count)
for _ in range(10000):
    i = np.random.choice(inices, p=(miners/np.sum(miners)))
    winners.append(i)
    miners[i] += 300
    miners = (miners>0)*miners
    miners += (miners>0)*-1
```

```python
plt.hist(miners)
plt.show()
```

```python
unique_counts = []
for s in np.split(np.array(winners),len(winners)/100):
    unique_counts.append(len(np.unique(s)))
```

```python
plt.plot(unique_counts)
plt.xlabel('Hundreds of blocks', fontsize=14)
plt.ylabel('Unique Miners', fontsize=15)

plt.show()
```
